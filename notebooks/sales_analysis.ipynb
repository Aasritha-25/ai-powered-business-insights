# AI-Powered Business Insights
## Sales Data Analysis

This notebook performs exploratory data analysis on sales data and prepares insights for AI-based summary generation.

Stage 1: Load Data

```python
import pandas as pd

url = "https://raw.githubusercontent.com/Aasritha-25/ai-powered-business-insights/main/data/sales_data.csv"
df = pd.read_csv(url)
df
```

Stage 2: Understand the Data(EDA - Exploratory Data Analysis)

```python
df.info()
df.describe()
```

Stage 3: Exploratory Business Insights


```python
total_sales = df['sales'].sum()
total_profit = df['profit'].sum()
print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
sales_by_region = df.groupby('region')['sales'].sum()
sales_by_region
```

```python
profit_by_product = df.groupby('product')['profit'].sum()
profit_by_product
```

Stage 4: Visualization

```python
import matplotlib.pyplot as plt

sales_by_region.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

Stage 5: Insights  
            
## Business Insights

- South region generates the highest sales.
- Electronics category contributes more revenue than Furniture.
- Laptops are the most profitable product.
- West region has high-value transactions.

These insights can help management improve regional sales strategies and inventory planning.
